```python
import os
import json

# Specify the folder path
root_folder_path = "data/"

# Initialize a list to store the modified data
modified_data = []

# Iterate over all folders in the root directory
for folder_name in os.listdir(root_folder_path):
    folder_path = os.path.join(root_folder_path, folder_name)

    # Check if the path is a directory
    if os.path.isdir(folder_path):
        # Iterate over all JSON files in the folder
        for filename in os.listdir(folder_path):
            if filename.endswith('.json'):
                # The label is the folder_name
                label = folder_name
                
                # Open the JSON file
                with open(os.path.join(folder_path, filename), 'r') as file:
                    data = json.load(file)
                
                # Append the 'label' tuple to each object in the JSON data
                for obj in data:
                    obj['label'] = label
                
                # Append the modified data to the list
                modified_data.extend(data)
```

```python
import pandas as pd

# Create a list to store the extracted data
extracted_data = []

# Iterate over the modified data
for obj in modified_data:
    attributes = list(obj.values())
    if len(attributes) >= 2:
        extracted_data.append({
            'text': attributes[1],
            'label': obj['label']
        })

# Create a DataFrame from the extracted data
df = pd.DataFrame(extracted_data)
```

```python
# Define the mapping of old labels to new labels
label_mapping = {
    'facebook': 'facebook_posts',
    'rmrj': 'rmrj_articles',
    'religious': 'religious_admin',
    'usjr': 'usjr_about'
}

# Update the values in the 'label' column
df['label'] = df['label'].replace(label_mapping)

# Print the updated DataFrame
print(df)
```
