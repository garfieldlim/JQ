```python
import pandas as pd
import openai

# Set up OpenAI API credentials
openai.api_key = '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'

def generate_description(department):
    prompt = f"These are all departments of USJ-R (University of San Jose- Recoletos). You are tasked to give a description of the contacts of such department. Department: {department}. Description: "
    response = openai.Completion.create(
        engine='text-davinci-003',
        prompt=prompt,
        max_tokens=200,
        n=1,
        stop=None,
        temperature=0.7
    )
    return response.choices[0].text.strip()

# Add 'text' column with generated descriptions
df['text'] = df['Department'].apply(generate_description)
```
