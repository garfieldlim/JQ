```python
import uuid

def split_text(item, max_length=4500):
    text = item['text']
    if len(text) <= max_length:
        return [item]

    parts = []
    for i in range(0, len(text), max_length):
        new_item = item.copy()
        new_item['text'] = text[i:i+max_length]
        parts.append(new_item)
    return parts

# Flatten the list of lists and sort by 'text_id'
all_collections['text'] = sorted(
    [part for item in all_collections['text'] for part in split_text(item)], 
    key=lambda x: x['text_id']
)
```
