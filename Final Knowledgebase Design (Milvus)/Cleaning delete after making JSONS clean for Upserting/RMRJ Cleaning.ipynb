Splitting

```python
def split_text(data):
    new_data = []
    for item in data:
        text = item['text']
        if len(text) > 4500:
            chunks = [text[i:i+4500] for i in range(0, len(text), 4500)]
            for chunk in chunks:
                new_item = item.copy()
                new_item['text'] = chunk
                new_data.append(new_item)
        else:
            new_data.append(item)
    return new_data

# Usage:
all_collections['text'] = split_text(all_collections['text'])

for item in all_collections['text']:
    item['uuid'] = str(uuid.uuid4())
```
