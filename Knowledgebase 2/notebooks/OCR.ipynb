```python
from PIL import Image
import os
import pytesseract

source_dir = "/Users/garfieldgreglim/Documents/JQ/Knowledgebase 2/images"
output_dir = "/Users/garfieldgreglim/Documents/JQ/Knowledgebase 2/image_ocrs"

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for filename in os.listdir(source_dir):
    if filename.endswith(".jpg"):  # Process only .jpg files
        file_path = os.path.join(source_dir, filename)
        image = Image.open(file_path)

        # Perform OCR
        text = pytesseract.image_to_string(image)

        # Save text to a .txt file
        output_file = os.path.join(output_dir, os.path.splitext(filename)[0] + ".txt")
        with open(output_file, 'w', encoding='utf-8') as file:
            file.write(text)

        print(f"Processed {filename}")
```

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import os

source_dir = "/Users/garfieldgreglim/Documents/JQ/Knowledgebase 2/pdfs"
output_dir = "/Users/garfieldgreglim/Documents/JQ/Knowledgebase 2/image_ocrs"

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for filename in os.listdir(source_dir):
    if filename.endswith(".pdf"):
        file_path = os.path.join(source_dir, filename)
        doc = fitz.open(file_path)

        all_text = ""

        for page_num in range(len(doc)):
            page = doc.load_page(page_num)  # number of page
            pix = page.get_pixmap()
            image = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

            # Perform OCR on the image
            text = pytesseract.image_to_string(image)
            all_text += text

        # Save text to a .txt file
        output_file = os.path.join(output_dir, os.path.splitext(filename)[0] + ".txt")
        with open(output_file, 'w', encoding='utf-8') as file:
            file.write(all_text)

        print(f"Processed {filename}")
```
